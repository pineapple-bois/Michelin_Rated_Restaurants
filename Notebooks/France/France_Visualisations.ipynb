## Michelin Stars by department and region

### `top_restaurants_by_department` function

Defined in `functions.py`, this function analyses a dataset of restaurants and outputs the top $n$ departments with the highest count of restaurants with a specified Michelin star rating.

Arguments:

- `data` (pandas.DataFrame): The dataset containing restaurant info.
- `star_rating` (int): The Michelin star rating (1, 2, or 3).
- `top_n` (int): The number of departments to consider.
- `display_restaurants` (bool): Whether to display individual restaurants. Default is True.
- `display_info` (bool): prints style of cuisine and URL. Default is False.

`top_restaurants_by_region` works in a similar way but accepts the `all` keyword for `star_rating` and `top_n`

#### By Department

```python
# excludes Paris
top_restaurants_by_department(rest_of_france, 3, 5, display_info=True)
```

```python
# We specify 'top5' however, the paris dataset has 1 department; 75
top_restaurants_by_department(paris, 3, 5, display_info=True)
```

#### By Region

```python
# For brevity, the display_restaurants flag is set False
# Using 'all' parameter for star_rating
top_restaurants_by_region(rest_of_france, 'all', 3, display_restaurants=False)
```

### `plot_choropleth` & `plot_choropleth_multi` functions

Arguments:

- `df` (GeoDataFrame): The DataFrame containing the data.
- `column` (str): The name of the column to plot.
- `title` (str): The title of the plot.
- `regional` (bool): If True, show regional map with departments labeled. Default is False.
- `restaurants` (bool): If True, plot restaurant locations. Default is False.
- `cmap` (str): The colormap to use. Default is 'Blues'. [matplotlib colour maps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)
- `figsize` (tuple): The size of the figure. Default is (10, 10).

Lists of `column` and `title` are passed to the `plot_choropleth_multi` function

```python
plot_choropleth(no_paris, 'total_stars', "Michelin Stars by Department\n(Excluding Paris)")
```

```python
plot_choropleth(no_paris, 'total_restaurants', "Michelin Restaurants by Department\n(Excluding Paris)")
```

```python
plot_choropleth(no_paris, '3_star', "Three Star Restaurants by Department\n(Excluding Paris)")
```

```python
plot_choropleth(no_paris, '2_star', "Two Star Restaurants by Department\n(Excluding Paris)")
```

```python
plot_choropleth(no_paris, '1_star', "One Star Restaurants by Department\n(Excluding Paris)")
```
